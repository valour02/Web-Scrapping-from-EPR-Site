```python
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from tqdm import tqdm
import pandas as pd
```

```python
path = "C:\\chromedriver.exe"
service = Service(executable_path = path)
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)

link = 'https://eprplastic.cpcb.gov.in/#/plastic/home/main_dashboard'
driver.get(link)

wait = WebDriverWait(driver, 10)
collapsable_element = wait.until(EC.element_to_be_clickable(
    (By.XPATH, '/html/body/app-root/app-plastic/div/app-main-dashboard/div/div/div[3]/div[3]/div[1]/div/div[2]/div[1]/div[2]/div[1]/span')))
collapsable_element.click()

driver.maximize_window()
sleep(3)

tbody=driver.find_element(By.XPATH,'/html/body/ngb-modal-window/div/div/div[2]/kl-simple-table-with-pagination/div[1]/div/div[1]/table/tbody')

BO_InProcess= []
for tr in tqdm(tbody.find_elements(By.XPATH,'//tr')):
    row = [item.text for item in tr.find_elements(By.XPATH,'.//td')]
    BO_InProcess.append(row)


BoInProcess = pd.DataFrame(BO_InProcess, columns = ['Sno.','Name','Address','',''])
BoInProcess = BoInProcess.dropna(axis=0)


BoInProcess.to_csv('C:\EPR_Scrapping\EPR_Plastic_Data_Brand_Owners_InProcess.csv',index=False)
```
